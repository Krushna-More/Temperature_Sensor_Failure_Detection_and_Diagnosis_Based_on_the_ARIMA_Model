```python
# X1 = thermistor
# X2 = thermocouple
# X3 = rtd
# X4 = lm35
X1 = thermistor.values
train1 = X1[0:149] # 149 data as train data
test1 = X1[150:]  # 50 data as test data
predictions1 = []
print(train1.size)
print(test1.size)
print("\n")

X2 = thermistor.values
train2 = X2[0:149] # 149 data as train data
test2 = X2[150:]  # 50 data as test data
predictions2 = []
print(train2.size)
print(test2.size)
print("\n")

X3 = thermistor.values
train3 = X3[0:149] # 149 data as train data
test3= X3[150:]  # 50 data as test data
predictions3 = []
print(train3.size)
print(test3.size)
print("\n")

X4 = thermistor.values
train4 = X4[0:149] # 149 data as train data
test4 = X4[150:]  # 50 data as test data
predictions4 = []
print(train4.size)
print(test4.size)
```

```python
from statsmodels.tsa.ar_model import AR
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
model_ar1 = AR(train1)
model_ar_fit1 = model_ar1.fit()
predictions1 = model_ar_fit1.predict(start=149,end=200)

model_ar2 = AR(train2)
model_ar_fit2 = model_ar2.fit()
prediction2 = model_ar_fit2.predict(start=149,end=200)

model_ar3 = AR(train3)
model_ar_fit3 = model_ar3.fit()
predictions3 = model_ar_fit3.predict(start=149,end=200)

model_ar4 = AR(train4)
model_ar_fit4 = model_ar4.fit()
predictions4 = model_ar_fit4.predict(start=149,end=200)
```

```python
#p,d,q  p = periods taken for autoregressive model
#d -> Integrated order, difference
# q periods in moving average model
model_arima1 = ARIMA(train1,order=(4, 2, 4))
model_arima_fit1 = model_arima1.fit()
print(model_arima_fit1.aic) #lesser the aic values better is the model

model_arima2 = ARIMA(train2,order=(4, 2, 4))
model_arima_fit2 = model_arima2.fit()
print(model_arima_fit2.aic) #lesser the aic values better is the model

model_arima3 = ARIMA(train3,order=(4, 2, 4))
model_arima_fit3 = model_arima3.fit()
print(model_arima_fit3.aic) #lesser the aic values better is the model

model_arima4 = ARIMA(train4,order=(4, 2, 4))
model_arima_fit4 = model_arima4.fit()
print(model_arima_fit4.aic) #lesser the aic values better is the model
```

```python
predictions1= model_arima_fit1.forecast(steps=50)[0]
print(predictions1)

predictions2= model_arima_fit2.forecast(steps=50)[0]
print(predictions2)

predictions3= model_arima_fit3.forecast(steps=50)[0]
print(predictions3)

predictions4= model_arima_fit4.forecast(steps=50)[0]
print(predictions4)
```
